```python
train_copy=train_data.copy()
train_copy.reset_index(drop=True,inplace=True)
```

```python
import numpy as np
import pandas as pd

# Example: using train_copy as a DataFrame
# train_copy = pd.DataFrame(...)

# Function to get outliers using IQR
def get_outliers(X):
    Q1 = np.quantile(X, 0.25)
    Q3 = np.quantile(X, 0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_indices = np.where((X < lower_bound) | (X > upper_bound))[0]
    outliers = X[outlier_indices]

    return outliers, outlier_indices.tolist()

# Dictionary to store outliers for each feature
all_outliers = {}

for col in train_copy.columns:
    outliers, outlier_indices = get_outliers(train_copy[col])
    all_outliers[col] = {
        "outliers": outliers,
        "indices": outlier_indices,
        "count": len(outlier_indices)
    }

# Display outliers info
for feature, info in all_outliers.items():
    print(f"Feature: {feature}")
    print(f"Number of outliers: {info['count']}")
    print(f"Indices of outliers: {info['indices']}")
    print(f"Outlier values: {info['outliers'].tolist()}\n")
```

```python
all_outlier_indices = set()
for info in all_outliers.values():
    all_outlier_indices.update(info['indices'])
```

```python
train_copy.drop(all_outlier_indices,inplace=True)
train_copy.reset_index(drop=True,inplace=True)
```
